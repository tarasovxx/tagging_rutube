```python
from handler import convert_mp4_to_txt
import glob
import os

folder_path = "videos"

video_files = glob.glob(os.path.join(folder_path, "*.mp4"))
results = []


for i, video_file in enumerate(video_files):
        data = convert_mp4_to_txt(video_file)
        print(f"{i}, {data[:10]}")
        results.append({"video_id": video_file.split('/')[1].split('.mp4')[0], "transcribe": data})
df = pd.DataFrame(results)
df.head()
```

```python
df.to_csv('video_with_transcribe.csv', index=False)
```
